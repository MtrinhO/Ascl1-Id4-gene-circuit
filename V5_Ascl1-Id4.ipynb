```python
#Demo array input, output array
import numpy as np

def quiescense(Genearray=np.random.rand(2),lesser_weight=0.99, pos_bias=1.0):
    #Independant variables: Id4, Huwe1. Huwe1 effect dilluted by factor of 10
    beta_Id4=Genearray[0]
    beta_Huwe1=Genearray[1]
    Modified_Genearray = []
    #Calculates non_Inputted gene expressions E47 and Ascl1, and quiescence state
    beta_E47 = pos_bias - beta_Id4
    beta_Ascl1 = beta_E47*(beta_Huwe1*lesser_weight)
    #Returns sample array with decisions made according to circuit
    Modified_Genearray.extend([beta_Id4, beta_Huwe1, beta_E47, beta_Ascl1])
    verdict = decision(Modified_Genearray)
    Modified_Genearray.append(verdict)
    return Modified_Genearray

#Classifies quiescence value as indicating an active, quiescent, or abnormally active stem cell
def decision(Modified_Genearray):
    a=Modified_Genearray[3]
    verdict=np.where((a<0.4), 2.0, 3.0)
    return verdict

#KEY: QUIESCENT = 2.0  ACTIVE = 3.0  HYPERACTIVE = 4.0
```
